```python
import pandas as pd
from PIL import Image

df_train = pd.read_csv('mnist-pngs/train.csv')
df_test = pd.read_csv('mnist-pngs/test.csv')
# df_train.head()
df_test.head()
```

```python
# create validation set
df_val = df_train.sample(frac=0.2, random_state=42)
df_train = df_train.drop(df_val.index)

df_train.reset_index(drop=True, inplace=True)
df_val.reset_index(drop=True, inplace=True)
```

```python
#save to csv
df_train.to_csv('mnist-pngs/new_train.csv', index=False)
df_val.to_csv('mnist-pngs/new_val.csv', index=False)
```

```python
from torch.utils.data import Dataset, dataloader
import os

class LoadDataset(Dataset):
    def __init__(self, csv_path, img_dir, transform=None) -> None:
        super().__init__()
        self.df = pd.read_csv(csv_path)
        self.img_dir = img_dir
        self.transform = transform
        # print("IMAGE", self.img_dir)y
        

    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx):
        img_name = os.path.join(self.img_dir, self.df.iloc[idx, 0])
        image = Image.open(img_name).convert("RGB")
        label = self.df.iloc[idx, 1]
        if self.transform is not None:
            image = self.transform(image)
        return image, label


```

```python
# defing transforms

from torchvision import transforms

data_transform = {
    "train": transforms.Compose(
        [
            transforms.RandomCrop(28, padding=4),
            transforms.RandomHorizontalFlip(),
            transforms.ToTensor(),
            transforms.Normalize((0.1307,), (0.3081,)),
        ]
    ),
    "test": transforms.Compose(
        [
            transforms.ToTensor(),
            transforms.Normalize((0.1307,), (0.3081,)),
        ]
    ),

}
```

```python
# define dataloader

from torch.utils.data import DataLoader

train_dataset = LoadDataset(
    csv_path="mnist-pngs/new_train.csv",
    img_dir="mnist-pngs/",
    transform=data_transform["train"],
)

train_loader = DataLoader(
    dataset=train_dataset,
    batch_size=64,
    shuffle=True,
    num_workers=4,

)

test_dataset = LoadDataset(
    csv_path="mnist-pngs/test.csv",
    img_dir="mnist-pngs/",
    transform=data_transform["test"],

)

test_loader = DataLoader(
    dataset=test_dataset,
    batch_size=64,
    num_workers=4,
)

val_dataset = LoadDataset(
    csv_path="mnist-pngs/new_val.csv",
    img_dir="mnist-pngs/",
    transform=data_transform["train"],
)

val_loader = DataLoader(
    dataset=val_dataset,
    batch_size=64,
    num_workers=4,
)

```
